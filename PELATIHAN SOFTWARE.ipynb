```python
data_asuransi_drop['PENGALAMAN_MENGEMUDI'] = \
data_asuransi_drop['PENGALAMAN_MENGEMUDI'].map({"lebih dari 30 tahun":3,"antara 20-30 tahun":2,"antara 10-20 tahun":1, "kurang dari 10 tahun":0})
```

```python
data_asuransi_drop['PENGHASILAN'] = \
data_asuransi_drop['PENGHASILAN'].map({"lebih dari 3x UMR":3,"antara 2-3x UMR":2,"antara 1-2x UMR":1, "kurang dari UMR":0})
```

```python
data_asuransi_drop['TAHUN_PRODUKSI'] = \
data_asuransi_drop['TAHUN_PRODUKSI'].map({"setelah 2020":1,"sebelum 2020":0})
```

```python
data_asuransi_drop['KLAIM'] = \
data_asuransi_drop['KLAIM'].map({"tidak":1,'ya':0})
```

```python
data_categorical = ['JENIS_KELAMIN','PEMILIK_KENDARAAN',"STATUS","ANAK","JENIS_KENDARAAN"]
data_numeric = ['UMUR',"PENDIDIKAN",'PENGALAMAN_MENGEMUDI',"PENGHASILAN"]
data_binary = ['TAHUN_PRODUKSI']
data_target = ['KLAIM']
```

### One Hot Encoding

```python
# One Hot Encoding for categorical data
data_categorical = pd.get_dummies(data_asuransi)
data_categorical.head()
```

```python
# Drop data bentuk numeric karena akan dinormalisasi dulu
data_asuransi_drop = data_categorical.drop(data_numeric, axis =1) 
```

```python
data_asuransi_drop.head()
```

```python
data_asuransi[data_numeric].describe()
```

### Menggunakan Normalisasi

```python
# Normalisasi data dalam bentuk numerik
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(data_asuransi_drop[data_numeric])
data_normalisasi = pd.DataFrame(d, columns=data_asuransi_drop[data_numeric].columns)
data_normalisasi.head(10)
```

```python
x = data_asuransi_drop.drop(['KLAIM'], axis = 1)
y = data_asuransi_drop['KLAIM']
```

### Membagi data menjadi data training dan testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

### Menggunakan Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model_svm = SVC()
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_test,y_test)
```

### KFold SVM

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
score_svm = cross_val_score(model_svm,x,y,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score_svm)
```
